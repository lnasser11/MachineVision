```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernel = np.array([ [0, 1, 0], [1, -4, 1], [0, 1, 0] ], dtype = 'int16')

img1 = cv2.imread('Figuras/Fig0338a.tif', cv2.IMREAD_GRAYSCALE)

(h, w) = img1.shape
m = 3

d = int((m-1)/2)
fig_out = np.zeros((h, w), dtype = 'uint8')

for i in range(d, h-d):
    for j in range(d, w-d):
        
        fig_section = img1[i - d : i + d + 1, j - d : j + d + 1] * kernel

        fig_out[i, j] = abs(np.sum(fig_section))



fig_final = img1.astype('int16') + fig_out.astype('int16')

fig_final = np.clip(fig_final, 0, 255)
fig_final = fig_final.astype('uint8')

# plot all images side by side
plt.figure(figsize=(10, 10))
plt.subplot(131), plt.imshow(img1, cmap='gray'), plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(fig_out, cmap='gray'), plt.title('Filtered Image')
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(fig_final, cmap='gray'), plt.title('Final Image')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img_panda1 = cv2.imread('Figuras/Onde_Esta_Panda1_U8.bmp', cv2.IMREAD_GRAYSCALE)
panda = cv2.imread('Figuras/Panda_Padrao_u8.bmp', cv2.IMREAD_GRAYSCALE)

(h, w) = img_panda1.shape
print(h, w)

(m, n) = panda.shape
print(m, n)

d1 = int((m-1)/2)
d2 = int((n-1)/2)

s_min = 2000000000
sad = np.ones((h,w), dtype='int32') * s_min

for i in range(d1, h - d1):
    for j in range(d2, w - d2):

        sec_img = img_panda1[i - d1 : i + d1 + 1, j - d2 : j + d2 + 1]
        mat_dif = abs(sec_img.astype(np.int32) - panda.astype(np.int32))
        sad[i,j] = np.sum(mat_dif)

        #achar o menor index dentro da varredora do for
        if (sad[i,j] < s_min):
            i_min = i
            j_min = j
            s_min = sad[i,j]

print('Pos X = ', i_min)
print('Pos Y = ', j_min)
print('S_Min = ', s_min)

img_end = cv2.cvtColor(img_panda1, cv2.COLOR_GRAY2BGR)
img_end_ret = cv2.rectangle(img_end, (j_min - d2, i_min - d1), (j_min + d2, i_min + d1), (255, 0, 0), 2)

plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(img_panda1, cmap='gray'), plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img_end_ret), plt.title('Final Image')
plt.xticks([]), plt.yticks([])
plt.show()
```
