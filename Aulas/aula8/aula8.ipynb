## Transformação Geométrica

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('Figuras/screw_u8.bmp', cv2.IMREAD_GRAYSCALE)
(h, w) = img1.shape
```

```python

# Escala
TH1 = np.array( [[2, 0, 0],
                [0, 2, 0],
                [0, 0, 1]], dtype= np.float32)

img_out1 = np.ones((h,w), dtype=np.uint8) * 255

# Translação
TH2 = np.array( [[1, 0, 300],
                [0, 1, 100],
                [0, 0, 1]], dtype= np.float32)

img_out2 = np.ones((h,w), dtype=np.uint8) * 255

# Cisalhamento
TH3 = np.array( [[1, 1, 0],
                [0, 1, 0],
                [0, 0, 1]], dtype= np.float32)

img_out3 = np.ones((h,w), dtype=np.uint8) * 255

# Rotação 20°
TH4 = np.array( [[np.cos(np.radians(20)), -np.sin(np.radians(20)), 0],
                [np.sin(np.radians(20)), np.cos(np.radians(20)), 0],
                [0, 0, 1]], dtype= np.float32)

img_out4 = np.ones((h,w), dtype=np.uint8) * 255

for x in range(w):
    for y in range(h):

        p0_1 = np.array([ [x], [y], [1] ])
        p1_1 = np.matmul(TH1, p0_1)
        u_1 = int(p1_1[0,0])
        v_1 = int(p1_1[1,0])

        if u_1 >= 0 and u_1 < w and v_1 >= 0 and v_1 < h:
            img_out1[v_1,u_1] = img1[y,x]

for x in range(w):
    for y in range(h):

        p0_2 = np.array([ [x], [y], [1] ])
        p1_2 = np.matmul(TH2, p0_2)
        u_2= int(p1_2[0,0])
        v_2= int(p1_2[1,0])

        if u_2 >= 0 and u_2 < w and v_2 >= 0 and v_2 < h:
            img_out2[v_2,u_2] = img1[y,x]

for x in range(w):
    for y in range(h):

        p0_3 = np.array([ [x], [y], [1] ])
        p1_3 = np.matmul(TH3, p0_3)
        u_3= int(p1_3[0,0])
        v_3= int(p1_3[1,0])

        if u_3 >= 0 and u_3 < w and v_3 >= 0 and v_3 < h:
            img_out3[v_3,u_3] = img1[y,x]

for x in range(w):
    for y in range(h):

        p0_4 = np.array([ [x], [y], [1] ])
        p1_4 = np.matmul(TH4, p0_4)
        u_4= int(p1_4[0,0])
        v_4= int(p1_4[1,0])

        if u_4 >= 0 and u_4 < w and v_4 >= 0 and v_4 < h:
            img_out4[v_4,u_4] = img1[y,x]

plt.figure(figsize=(20, 10))

# Original Image
plt.subplot(1, 5, 1)  # 1 linha, 5 colunas, índice 1
plt.imshow(img1, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Aumento de Escala
plt.subplot(1, 5, 2)  # índice 2
plt.imshow(img_out1, cmap='gray')
plt.title('Aumento de Escala')
plt.axis('off')

# Translação
plt.subplot(1, 5, 3)  # índice 3
plt.imshow(img_out2, cmap='gray')
plt.title('Translação')
plt.axis('off')

# Cisalhamento
plt.subplot(1, 5, 4)  # índice 4
plt.imshow(img_out3, cmap='gray')
plt.title('Cisalhamento')
plt.axis('off')

# Rotação 20°
plt.subplot(1, 5, 5)  # índice 5
plt.imshow(img_out4, cmap='gray')
plt.title('Rotação 20°')
plt.axis('off')

plt.show()
```

```python
img2 = cv2.imread('Figuras/Fig_TH_U8.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
(h, w) = img2.shape

# Rotação
TH = np.array( [[np.cos(np.radians(20)), -np.sin(np.radians(20)), 0],
                [np.sin(np.radians(20)), np.cos(np.radians(20)), 0],
                [0, 0, 1]], dtype= np.float32)

img_out = np.ones((h,w), dtype=np.uint8) * 255


for x in range(w):
    for y in range(h):

        p0 = np.array([ [x], [y], [1] ])
        p1 = np.matmul(TH, p0)
        u = int(p1[0,0])
        v = int(p1[1,0])

        if u >= 0 and u < w and v >= 0 and v < h:
            img_out[v,u] = img2[y,x]

# Translação
TH = np.array( [[1, 0, 100],
                [0, 1, 50],
                [0, 0, 1]], dtype= np.float32)

img_out2 = np.ones((h,w), dtype=np.uint8) * 255

for x in range(w):
    for y in range(h):

        p0 = np.array([ [x], [y], [1] ])
        p1 = np.matmul(TH, p0)
        u = int(p1[0,0])
        v = int(p1[1,0])

        if u >= 0 and u < w and v >= 0 and v < h:
            img_out2[v,u] = img_out[y,x]

plt.figure(figsize=(20, 10))

plt.subplot(1, 3, 1)
plt.imshow(img2, cmap='gray')
plt.title('Imagem Original')

plt.subplot(1, 3, 2)
plt.imshow(img_out, cmap='gray')
plt.title('Rotação 20°')

plt.subplot(1, 3, 3)
plt.imshow(img_out2, cmap='gray')
plt.title('Rotação 20° + Translação')
```

```python
img3 = cv2.imread('Figuras/screw_u8.bmp', cv2.IMREAD_GRAYSCALE)

(h, w) = img3.shape

# Translação
TH = np.array( [[1, 0, 200],
                [0, 1, 100],
                [0, 0, 1]], dtype= np.float32)

img_out = np.ones((h,w), dtype=np.uint8) * 255

for x in range(w):
    for y in range(h):

        p0 = np.array([ [x], [y], [1] ])
        p1 = np.matmul(TH, p0)
        u = int(p1[0,0])
        v = int(p1[1,0])

        if u >= 0 and u < w and v >= 0 and v < h:
            img_out[v,u] = img3[y,x]
    
# Rotação
TH = np.array( [[np.cos(np.radians(45)), -np.sin(np.radians(45)), 0],
                [np.sin(np.radians(45)), np.cos(np.radians(45)) , 0],
                [0                     , 0                      , 1]], dtype= np.float32)

img_out2 = np.ones((h,w), dtype=np.uint8) * 255

for x in range(w):
    for y in range(h):

        p0 = np.array([ [x], [y], [1] ])
        p1 = np.matmul(TH, p0)
        u = int(p1[0,0])
        v = int(p1[1,0])

        if u >= 0 and u < w and v >= 0 and v < h:
            img_out2[v,u] = img_out[y,x]

# Translação
TH = np.array( [[1, 0, 10],
                [0, 1, -255],
                [0, 0, 1]], dtype= np.float32)

img_out3 = np.ones((h,w), dtype=np.uint8) * 255

for x in range(w):
    for y in range(h):

        p0 = np.array([ [x], [y], [1] ])
        p1 = np.matmul(TH, p0)
        u = int(p1[0,0])
        v = int(p1[1,0])

        if u >= 0 and u < w and v >= 0 and v < h:
            img_out3[v,u] = img_out2[y,x]
    


plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.imshow(img3, cmap='gray')
plt.title('Imagem Original')

plt.subplot(1, 2, 2)
plt.imshow(img_out3, cmap='gray')
plt.title('Rotação 45° em relação ao próprio centro')
```
