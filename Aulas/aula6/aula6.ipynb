```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = cv2.imread('Figuras/Fig0335a.tif', cv2.IMREAD_GRAYSCALE) 

(h, w) = fig.shape

print('Height: ', h)
print('Width: ', w) 
```

```python
# Construindo o Kernel

m = 3

d = int((m-1)/2)
kernel = np.ones((m,m), dtype = 'float') / m**2
print(kernel)

fig_out = np.zeros((h, w), dtype = 'uint8')

for i in range(d, h-d):
    for j in range(d, w-d):
        
        fig_out[i, j] = np.sum(fig[i-d:i+d+1, j-d:j+d+1] * kernel)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(fig, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(fig_out, cmap='gray')
plt.title('Filtrada')
plt.axis('off')

plt.show()
```

```python
# Construindo o Kernel

m = 3

d = int((m-1)/2)
kernel = np.ones((m,m), dtype = 'float') / m**2
print(kernel)

fig_out = np.zeros((h, w), dtype = 'uint8')

for i in range(d, h-d):
    for j in range(d, w-d):
        
        fig_section = fig[i-d:i+d+1, j-d:j+d+1] * kernel

        fig_out[i, j] = np.median(fig_section)



plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(fig, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(fig_out, cmap='gray')
plt.title('Filtrada')
plt.axis('off')

plt.show()
```

```python
fig_2 = cv2.imread('Figuras/Fig0342a.tif', cv2.IMREAD_GRAYSCALE)

m = 3
d = int((m-1)/2)

kernel_x = np.array([ [-1, -1, -1], [0, 0, 0], [1, 1, 1] ], dtype = 'int16')
kernel_y = np.array([ [-1, 0, 1], [-1, 0, 1], [-1, 0, 1] ], dtype = 'int16')

(h, w) = fig_2.shape

plt.imshow(fig_2, cmap='gray')
plt.axis('off')
```

```python
fig_out_x = np.zeros((h, w), dtype = 'uint8')

for i in range(d, h-d):
    for j in range(d, w-d):
        
        fig_section = fig_2[i-d:i+d+1, j-d:j+d+1] * kernel_x

        fig_out_x[i, j] = abs(np.sum(fig_section))
```

```python
plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(fig_2, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(fig_out_x, cmap='gray')
plt.title('Filtrada X')
plt.axis('off')

plt.show()
```

```python
fig_out_y = np.zeros((h, w), dtype = 'uint8')

for i in range(d, h-d):
    for j in range(d, w-d):
        
        fig_section = fig_2[i-d:i+d+1, j-d:j+d+1] * kernel_y

        fig_out_y[i, j] = abs(np.sum(fig_section))
```

```python
plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(fig_2, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(fig_out_y, cmap='gray')
plt.title('Filtrada Y')
plt.axis('off')

plt.show()
```

```python
fig_filtrada_final = fig_out_x + fig_out_y

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(fig_2, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(fig_filtrada_final, cmap='gray')
plt.title('Filtrada Final')
plt.axis('off')

plt.show()
```

```python
# Cdigo Kernels X e Y juntos para chegar na figura final direto

fig_2 = cv2.imread('Figuras/Fig0342a.tif', cv2.IMREAD_GRAYSCALE)

m = 3
d = int((m-1)/2)

kernel_x = np.array([ [-1, -1, -1], [0, 0, 0], [1, 1, 1] ], dtype = 'int16')
kernel_y = np.array([ [-1, 0, 1], [-1, 0, 1], [-1, 0, 1] ], dtype = 'int16')

(h, w) = fig_2.shape

fig_out = np.zeros((h, w), dtype = 'uint8')

for i in range(d, h-d):
    for j in range(d, w-d):
        
        fig_section_x = fig_2[i-d:i+d+1, j-d:j+d+1] * kernel_x
        fig_section_y = fig_2[i-d:i+d+1, j-d:j+d+1] * kernel_y

        fig_out[i, j] = abs(np.sum(fig_section_x)) + abs(np.sum(fig_section_y))

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(fig_2, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(fig_out, cmap='gray')
plt.title('Filtrada Final')
plt.axis('off')

plt.show()
```

```python
img2 = cv2.imread('Figuras/Formas_U8.bmp', cv2.IMREAD_GRAYSCALE)
m = 3
d= int((m-1)/2)
kernel_Y = np.array([[-1,-1,-1],[0,0,0],[1,1,1]],dtype="int16")
kernel_X = np.array([[1,0,-1],[1,0,-1],[1,0,-1]],dtype="int16")

(h, w) = img2.shape

fig_out = np.zeros((h, w), dtype = 'uint8')
for i in range(d, h-d):
    for j in range(d, w-d):
        fig_section = img2[i-d:i+d+1, j-d:j+d+1]

        prod = kernel_X * fig_section
        sum_x = np.sum(prod)

        prod = kernel_Y * fig_section
        sum_y = np.sum(prod)

        fig_out[i, j] = abs(sum_x) + abs(sum_y)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(img2, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(fig_out, cmap='gray')
plt.title('Filtrada Final')
plt.axis('off')

plt.show()




```
