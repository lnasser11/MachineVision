### Desafio 4
#### Calcular % de n√≥s/veios da madeira

```python
img8 = cv2.imread('Figuras/Fig5.bmp', cv2.IMREAD_COLOR)
```

```python
plt.imshow(cv2.cvtColor(img8, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

#### Transformando de cor para escala de cinza

```python
(h,w,c) = img8.shape

print('Height: ', h)
print('Width: ', w)

img_gray = np.zeros((h,w), dtype = "uint8")

for i in range(h):
    for j in range(w):
        img_gray[i,j] = img8[i,j,0]*0.07 + img8[i,j,1]*0.72 + img8[i,j,2]*0.21


plt.imshow(img_gray, cmap = 'gray')
```

#### Transformando de escala de cinza para preto e branco

```python
(h, w) = img_gray.shape

print('Height: ', h)
print('Width = ', w)

img_bin = np.zeros((h,w), dtype = "uint8")

threshold = 100

for i in range(h):
    for j in range(w):

        if img_gray[i,j] > threshold:
            img_bin[i,j] = 255

plt.figure("Fig Bin")
plt.imshow(img_bin, cmap='gray')
plt.show()
```

#### Calculando a porcentagem de n√≥s

```python
(h, w) = img_bin.shape

print('Height: ', h)
print('Width = ', w)

black = 0
total = h * w
for i in range(h):
    for j in range(w):

        if img_bin[i,j] == 0:
            black += 1
        

percent = (black / total) * 100

print('Pixels pretos:', black)
print('Total de pixels:', total)
print(f'% de n√≥s = {percent:.2f}%')
```
